## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{e} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
vector_e = [1, 1, 1, 8]
[i**2 for i in vector_e]
```

```
vector_e = [1, 1, 1, 8]
vector_f = [3, 3, 3, 3]

mag_e = np.sqrt(sum([i**2 for i in vector_e]))
mag_f = np.sqrt(sum([i**2 for i in vector_f]))

print('The magnitude of vector e is: ', mag_e)
print('The magnitude of vector f is: ', mag_f)

if mag_e > mag_f:
  print('Vector e is longer')
elif mag_e < mag_f:
  print('Vector f is longer')
else:
  print('They\'re the same length')
```

## 1.9 Show that the following vectors are orthogonal (perpendicular to each other):

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 0 \\ -1  \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix}
\end{align}

```
vector_g  = [1, 0, -1]
vector_h = [1, np.sqrt(2), 1]

sum([(vector_g[i]*vector_h[i]) for i in range(0, 3)])
```

$\vec{g} \cdot \vec{h} = 0 \therefore$ **$\vec{g}$ and $\vec{h}$ are orthogonal vectors**.

## 3.4 Find $HH^{-1}$ and then find $G^{-1}G$. Is $HH^{-1} == G^{-1}G$? Why or Why not?

```
h = np.array([[12, 11],
              [7, 10]])

h_inv = np.linalg.inv(h)

hh_inv = np.matmul(h, h_inv)

print(hh_inv)
```

```
j = np.array([[0, 1, 2],
         [7, 10, 4],
         [3, 2, 0]])

j_inv = np.linalg.inv(j)

jj_inv = np.matmul(j, j_inv)

print(jj_inv)
```
